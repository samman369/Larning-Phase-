### Measure of Central Tendency
1. mean
2. median
3. mode

```python
# using numpy
ages=[23,24,32,45,12,43,67,45,32,32,35,56,32,120]
import numpy as np
print(np.mean(ages))
print(np.median(ages))
```

```python
# using statistic
import statistics
print(statistics.mean(ages))
print(statistics.median(ages))
print(statistics.mode(ages))
```

```python
# find outlier using boxplot
import seaborn as sns
sns.boxplot(ages)
```

```python
# check outlier boundary
q1,q3=np.percentile(ages,[25,75])
print("First quantile is {} and second quartile is {}".format(q1,q3))
IQR=q3-q1
lower_fence=q1-1.5*(IQR)
higher_fence=q3+1.5*(IQR)
print(lower_fence,higher_fence)
```

### Measure of dispertion 
1. variance 
2. standard deviation 

```python
# using statistic function 
# it calculate sample variance (n-1)
statistics.variance(ages)
```

```python
# it calculate population variance 
statistics.pvariance(ages)
```

```python
# using numpy function
# it calculate population variance (n)
np.var(ages,axis=0) 
```

```python
# using function 
def variance(data,dof=0):
  
  ## mean of the data
  n=len(ages)
  mean=sum(data)/n
    
  ## variance
  deviation=[(x - mean) ** 2 for x in data]
  variance=sum(deviation)/(n-dof)
  return variance
```

```python
# dof means degree of freedom 
# 0 means population 1 means sample 
print(variance(ages,dof=1))
print(variance(ages,dof=0))
```

```python
# standard deviation 
import math
math.sqrt(statistics.pvariance(ages))
```

### Different types of distributed data

```python
# create a normal distributed data
s=np.random.normal(0.5,0.2,1000)
print(s)
```

```python
# show this normal distributed data 
sns.histplot(s,kde=True)
```

```python
# make a log normal dietribution 
# mean and standard deviation
mu, std = 3., 1. 
s = np.random.lognormal(mu, std, 1000)
```

```python
# show this log normal data through histogram
sns.histplot(s,kde=True)
```

```python
# convert it into normal distribution data 
sns.histplot(np.log(s),kde=True)
```

```python
# progame to check the data is normally distributed or not 
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
# create a normal distributed data
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
# we can also check log distribution function 
# just add an log function in front of data 
def plot_data_log(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(np.log(sample),dist='norm',plot=pylab)
    plt.show()
```

```python
# make a log normal dietribution 
# mean and standard deviation
mu, std = 3., 1. 
s = np.random.lognormal(mu, std, 1000)
```

```python
plot_data_log(s)
```
